```python
import math
```

```python
class Neuron():
    def __init__(self, x0, x1, x2, x3, w0, w1, w2, w3, yr, dd):
        # Навчальні приклади
        self.x0 = x0
        self.x1 = x1
        self.x2 = x2
        self.x3 = x3
        
        # Вагові коефіцієнти
        self.w0 = w0
        self.w1 = w1
        self.w2 = w2
        self.w3 = w3
        
        # Допустима похибка
        self.dd = dd
        
        self.yr = yr
        
        
    def learn(self):
        xs = 0
        for n in range(0, 4):
            xs += eval(f"self.x" + str(n)) * eval(f"self.w" + str(n))
        y = 1 / (1 + math.exp(-xs))
        dn = abs((self.yr - y) / self.yr)
        print("y: ",y)
        print("dn:",dn)
        if dn <= self.dd:
            print(self.w0, self.w1, self.w2, self.w3)
            return 0
        else:
            q = y * (1 - y) * (self.yr - y)
            dw0 = self.x0 * q
            dw1 = self.x1 * q
            dw2 = self.x2 * q
            dw3 = self.x3 * q

            self.w0 = self.w0 + dw0
            self.w1 = self.w1 + dw1
            self.w2 = self.w2 + dw2
            self.w3 = self.w3 + dw3
            return self.learn()
    
    def recognition(self, x0, x1, x2, x3):
        xs = 0
        for n in range(0, 4):
            xs += eval(f"x" + str(n)) * eval(f"self.w" + str(n))
        y = 1 / (1 + math.exp(-xs))
        dn = abs((self.yr - y) / self.yr)
        print("y: ",y)
        print("dn:",dn)

neur = Neuron(1, 3, 5, 7, 0.1, 0.2, 0.3, 0.4, 0.3, 0.1)
neur.learn()
```

```python
print(neur.recognition(1, 3, 5, 7))
print(neur.recognition(1, 6, 5, 7))
print(neur.recognition(1, 2, 9, 7))
```
